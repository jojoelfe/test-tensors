## Generate 3D Test Tensors

Let's use the test-tensors library to generate different 3D tensor patterns for visualization.

````python
# Generate 3D tensor patterns using the library
print("Generating 3D tensor data using test-tensors library...")

# 1. Cubic cross pattern (default size)
cross_default = generate_cross_3d()
print(f"Default cross shape: {cross_default.shape}")

# 2. Smaller cubic cross pattern  
cross_small = generate_cross_3d(24)
print(f"Small cross shape: {cross_small.shape}")

# 3. Rectangular cross pattern
cross_rect = generate_cross_3d((16, 24, 32))
print(f"Rectangular cross shape: {cross_rect.shape}")

# Display basic statistics
datasets = {
    "Cross (Default 64続)": cross_default, 
    "Cross (Small 24続)": cross_small, 
    "Cross (Rectangular)": cross_rect
}

for name, data in datasets.items():
    print(f"\n{name}:")
    print(f"  Shape: {data.shape}")
    print(f"  Value range: {data.min():.1f} to {data.max():.1f}")
    print(f"  Non-zero voxels: {np.count_nonzero(data)}")
    print(f"  Data type: {data.dtype}")
````

## Complete Visualization Dashboard

The library provides a comprehensive `visualize_3d_tensor()` function that creates all the visualization views automatically:
- Central slices through all three dimensions
- Maximum intensity projections  
- Interactive 3D isosurface rendering

````python
# 1. Visualize the small cubic cross pattern
print("1. Small Cubic Cross Pattern")
visualize_3d_tensor(
    cross_small,
    title="Small Cross Pattern (24続)",
    figsize=(14, 9),
    threshold=0.5,
    show_3d=True  # Set to False if running without display
)
````

# 2. Visualize the rectangular cross pattern  
print("2. Rectangular Cross Pattern")
visualize_3d_tensor(
    cross_rect,
    title="Rectangular Cross Pattern (162432)",
    figsize=(16, 10),
    threshold=0.5,
    show_3d=True,
    pitch=(0.8, 1.0, 1.2)  # Different voxel spacing for each dimension
)

````python
# 3. Visualize the default large cross pattern
print("3. Default Cross Pattern (64続)")
visualize_3d_tensor(
    cross_default,
    title="Default Cross Pattern (64続)",
    figsize=(16, 10),
    threshold=0.5,
    show_3d=True
)
````

## Customization Options

The `visualize_3d_tensor()` function provides several customization options:

````python
# Example of customization options
print("4. Customized Visualization Example")

# Create a custom tensor pattern  
custom_cross = generate_cross_3d((12, 20, 16))

# Visualize with custom settings
visualize_3d_tensor(
    custom_cross,
    title="Custom Cross Pattern with Different Settings",
    figsize=(18, 12),        # Larger figure size
    threshold=0.3,           # Lower threshold for 3D isosurface
    show_3d=True,           # Enable 3D viewer
    pitch=(2.0, 1.0, 0.5)   # Custom voxel spacing (anisotropic)
)

print("\\nAvailable parameters:")
print("- volume: 3D numpy array to visualize") 
print("- title: Custom title for the visualization")
print("- figsize: Figure size as (width, height)")
print("- threshold: Threshold for 3D isosurface generation")
print("- show_3d: Enable/disable interactive 3D viewer")
print("- pitch: Voxel spacing in (Z, Y, X) dimensions")
````
